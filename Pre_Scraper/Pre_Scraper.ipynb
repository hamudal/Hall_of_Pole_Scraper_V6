```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the URL to scrape
url = "https://www.eversports.de/a/l/poledance/kassel"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')

# Find all containers with class "MuiStack-root css-3p760y"
containers = soup.find_all('div', class_='MuiStack-root css-3p760y')

# Create empty lists to store data
names = []
ratings = []
services = []
addresses = []

# Iterate over each container and find the required information
for container in containers:
    # Find all li elements within the current container
    lis = container.find_all('li', class_='MuiBox-root css-6gjw2k')
    
    # Iterate over each li element and extract information
    for li in lis:
        # Name
        name = li.find('h2', class_='MuiBox-root css-1qm3mld').text.strip()
        names.append(name)
        
        # Rating
        rating_img = li.find('img', alt='rating')
        rating = rating_img['alt'] if rating_img else None
        ratings.append(rating)
        
        # Services
        service_tags = li.find('div', class_='MuiStack-root css-1x464v2').find_all('p', class_='MuiTypography-root MuiTypography-body1 css-6ik050')
        service_list = [tag.text.strip() for tag in service_tags]
        services.append(service_list)
        
        # Address
        address = li.find('div', class_='MuiStack-root css-1x27eqg').find('p', class_='MuiTypography-root MuiTypography-body1 css-6ik050').text.strip()
        addresses.append(address)

# Create a DataFrame
data = {
    'Name': names,
    'Rating': ratings,
    'Services': services,
    'Address': addresses
}
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```
